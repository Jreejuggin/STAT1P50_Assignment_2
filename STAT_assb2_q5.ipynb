```python
import pandas as pd
import matplotlib.pyplot as plt



data = pd.read_csv('student.csv')

data = data[['studytime', 'grade']]

data2 = data.groupby('studytime').mean().reset_index()

studytime = data2['studytime']
gradeavg = data2['grade']

#shows the study time compared to the average grade
plt.plot(studytime, gradeavg, 'o-', label = 'line1',color= 'purple') 
plt.legend()
plt.show()

```
