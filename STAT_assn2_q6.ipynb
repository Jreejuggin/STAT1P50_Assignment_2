```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler

data = pd.read_csv('crime.csv')
data = data[['ViolentCrimesPerPop', 'PctPopUnderPov']]

crime_list = np.array(data['ViolentCrimesPerPop']).reshape(-1,1)
popUnder = np.array(data['PctPopUnderPov']).reshape(-1,1)

#makes scatter plot with just the regular values
plt.scatter(crime_list, popUnder, label = 'samples', color = 'red')
plt.legend()
plt.show()

scaler = StandardScaler()
data2 = scaler.fit_transform(data)

crime_list2=[]
popUnder2 = []

#for loop adds the corresponding scalar value to its respective list
for a, b in data2:
    crime_list2.append(a)
    popUnder2.append (b)

#makes scatter plot with scalar values    
plt.scatter(crime_list2, popUnder2, label = 'samples0.2', color = 'red')
plt.legend()
plt.show()
```
